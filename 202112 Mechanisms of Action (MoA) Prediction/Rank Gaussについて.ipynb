数値を順位に変換
⇒順位を保ったまま、ガウス分布に変換する

予測モデルにニューラルネットを使用する際の特徴量変換の手法として、Min-Max Normalizationや標準化よりも優れた性能を発揮する

```
import numpy as np
import matplotlib.pyplot as plt
!pip install japanize_matplotlib
import japanize_matplotlib
from sklearn.preprocessing import QuantileTransformer

data=np.arange(0,30).reshape(30,1)
transformer=QuantileTransformer(n_quantiles=10,random_state=0, output_distribution="normal") #normalがガウス分布を指定、n_quantilesはよくわからん
transformer.fit(data)
result=transformer.transform(data)
```

```
plt.hist(data,label='元データ')
plt.hist(result,label='RankGauss処理後')
plt.legend()
plt.show()
```
